```python
import logging
logging.basicConfig(level=logging.DEBUG)

import itertools
import time

import pipecat.limit
import pipecat.record
import pipecat.source.time
```

```python
def broken_metronome():
    for i in itertools.count():
        yield {}
        time.sleep(i)
```

```python
pipeline = broken_metronome()
pipeline = pipecat.source.time.timestamp(pipeline)
pipeline = pipecat.limit.timeout(pipeline, timeout=pipecat.quantity(5, pipecat.units.seconds))

for record in pipeline:
    pipecat.record.dump(record)
```
